Importing Required Libraries

```python
# Import pandas
import pandas as pd
```

Reading Dataset

```python
# Read the dataset using pandas
df = pd.read_csv("govtech_readiness_dataset.csv")
print(df.head())
```

Analyzing The Data

```python
# Overview of the dataset's size and structure
# This returns - (number of rows, number of columns)
df.shape
```

```python
# Converts the column names of the DataFrame into a Python list
df.columns.tolist()
```

```python
# Convert score_columns to numeric
score_columns = [
    'Population_millions',
    'GTMI_Score_2025',
    'CGSI_CoreGovSystems_2025',
    'PSDI_ServiceDelivery_2025',
    'DCEI_CitizenEngagement_2025',
    'GTEI_Enablers_2025',
    'EGov_Rank_2024',
    'EGov_Index_2024',
    'Online_Service_Index_2024',
    'Human_Capital_Index_2024',
    'Telecom_Infra_Index_2024'
]

for column in score_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

print("Converted all score columns to numeric format")
```

```python
# Understand the dataset
df.info()
```

```python
#  Statistical summary of the DataFrame (Transpose)
df.describe().T
```

Based on statistical summary, DCEI has a standard deviation of 0.192, which is the highest among the four GTMI sub-dimensions.

Checking For The Duplicate Values

```python
# Returns the number of duplicate rows in the dataset
df.duplicated().sum()
```

 Checking Missing Values

```python
#Identify any gaps in our data
df.isnull().sum()
```

Handling Missing Values

```python
# Convert '-' into missing values
df = df.replace('-', pd.NA)

# No imputation performed
print("Missing values retained as NaN")
```

Check Countries with Missing GTMI Score

```python
# Countries with missing GTMI scores cannot be included in GTMI-related charts and calculations
missing_gtmi = df[df['GTMI_Score_2025'].isna()]
display(missing_gtmi[['Country']])
print("These countries will be excluded from GTMI charts in Power BI.")
```

Add Helper Columns

```python
# 1. Replaces short region codes with readable names
region_name_map = {
    'ECA': 'Europe & Central Asia',
    'EAP': 'East Asia Pacific',
    'NAM': 'North America',
    'MNA': 'Middle East & N. Africa',
    'LCR': 'Latin Am. & Caribbean',
    'SAR': 'South Asia',
    'AFR': 'Sub-Saharan Africa'
}
df['Region_Full'] = df['WB_Region'].map(region_name_map)
```

```python
# 2. Give number to order_level for easy sorting
income_order_map = {'H': 1, 'UM': 2, 'LM': 3, 'L': 4}
df['Income_Order'] = df['Income_Level'].map(income_order_map)
```

```python
# 3. Income Full Name for making it more readable
income_name_map = {'H': 'High income', 'UM': 'Upper-middle', 'LM': 'Lower-middle', 'L': 'Low income'}
df['Income_Full'] = df['Income_Level'].map(income_name_map)
```

```python
print("Added 3 helper columns: Region_Full, Income_Order, Income_Full")
```

Save Clean File

```python
# Save a new CSV file in the same folder 
df.to_csv('govtech_clean_dataset.csv', index=False)
```
